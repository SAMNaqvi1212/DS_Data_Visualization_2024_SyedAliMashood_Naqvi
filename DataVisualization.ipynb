```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Pak_Climate_Change_Indicators_By_HDX.csv')
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Filter years for cleaner plots
years = df['Year']
```

```python
# CO2 Emissions from different sources (solid fuel, liquid fuel)
plt.figure(figsize=(12, 6))
plt.plot(years, df['CO2 emissions from solid fuel consumption (kt) Value'], label='Solid Fuel', color='red')
plt.plot(years, df['CO2 emissions from liquid fuel consumption (kt) Value'], label='Liquid Fuel', color='blue')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (kt)')
plt.title('CO2 Emissions from Different Sources Over Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Population Growth trends
plt.figure(figsize=(12, 6))
plt.plot(years, df['Population growth (annual %) Value'], label='Population Growth (Annual %)', color='green')
plt.plot(years, df['Urban population growth (annual %) Value'], label='Urban Population Growth (Annual %)', color='orange')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.title('Population Growth Trends Over Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(years, df['Population growth (annual %) Value'], color='blue', label='Population Growth (Annual %)')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.title('Annual Population Growth Over Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Scatter Plot: Annual Population Growth vs CO2 Emissions
plt.figure(figsize=(12, 6))
plt.scatter(df['Population growth (annual %) Value'], 
            df['CO2 emissions from solid fuel consumption (kt) Value'],
            label='CO2 Emissions vs Urban Population Growth',
            color='teal')
plt.xlabel('Annual Population Growth (Annual %)')
plt.ylabel('CO2 Emissions from Solid Fuel (kt)')
plt.title('Annual Population Growth vs CO2 Emissions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Rainfall Patterns
plt.figure(figsize=(12, 6))
plt.plot(years, df['RainFall Mean ()'], label='Mean Annual Rainfall', color='blue')
plt.xlabel('Year')
plt.ylabel('Rainfall (mm)')
plt.title('Mean Annual Rainfall Over Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Cereal Yield trends
plt.figure(figsize=(12, 6))
plt.plot(years, df['Cereal yield (kg per hectare) Value'], label='Cereal Yield (kg/ha)', color='purple')
plt.xlabel('Year')
plt.ylabel('Cereal Yield (kg/ha)')
plt.title('Cereal Yield Over Years')
plt.legend()
plt.grid(True)
plt.show()
```
